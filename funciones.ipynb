```
def sigmoid(a):
    return 1 / (1 + np.exp(-a))
```

```
def tanh(a):
    #return(np.exp(a) - np.exp(-a)) / (np.exp(a) + np.exp(-a))
    return np.tanh(a)
```

```
def softmax(v):
    v_exp = np.exp(v)
    v_exp_sum = np.sum(v_exp)
    return [a/v_exp_sum for a in v_exp]
```
